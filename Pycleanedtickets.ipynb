```python
date_cols = [
    "Created time", 
    "Expected SLA to resolve",
    "Expected SLA to first response", 
    "First response time",
    "Resolution time", 
    "Close time"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
#Creating First Response time in hours
df["FRT_hours"] = (
df["First response time"] - df["Created time"]
).dt.total_seconds() / 3600
```

```python
#Checking for any negative numbers
negatives = df[df["FRT_hours"] < 0]
print(negatives.shape)
print(negatives[["Ticket ID", "Created time",
"First response time", "FRT_hours"]])
```

```python
df["FRT_hours"].describe()
df.sort_values("FRT_hours").head(10)
df.sort_values("FRT_hours", ascending=False).head(10)
```

```python
#I first quantified how many there were and inspected them, so I could justify the decision rather than silently dropping data
before = df.shape[0]
df = df[df["FRT_hours"] >= 0]
after = df.shape[0]
print(f"Removed {before - after} invalid rows")
```

```python
df.describe()                 
df["FRT_hours"].describe() 
df["Priority"].value_counts() 
```

```python
df["created_hour"] = df["Created time"].dt.hour
```

```python
df["res_hours"] = (
df["Resolution time"] - df["Created time"]
).dt.total_seconds() / 3600
```

```python
df["created_weekday"] = df["Created time"].dt.day_name()
```
